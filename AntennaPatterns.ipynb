```python
import numpy as np
from pycbc.detector import Detector
import matplotlib.pyplot as plt
import os

n = 1000000
ra = np.random.uniform(0, 2 * np.pi, n)
dec = np.pi/2 - np.arccos(np.random.uniform(-1, 1, n))

pol = 0
t = 1e10

rho_min = 8

d = Detector("L1")

# Antenna pattern
fp, fc = d.antenna_pattern(ra, dec, pol, t)

P = fp**2 + fc**2
```

## Antenna pattern

```python
folder = "figures"
os.makedirs(folder, exist_ok=True)

x = P * np.cos(dec) * np.cos(ra)
y = P * np.cos(dec) * np.sin(ra)
z = P * np.sin(dec)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=P, cmap='viridis', s=2)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

cbar = plt.colorbar(sc, ax=ax, shrink=0.4, pad=0.1)
cbar.set_label('Antenna pattern')
plt.tight_layout()
ax.view_init(elev=30, azim=50)
file_path = os.path.join(folder, "antenna_pattern.pdf")
plt.savefig(file_path, format='pdf', bbox_inches='tight') 

plt.show()
plt.close()



```

## Amplitude antenna pattern

```python

fp, fc = d.antenna_pattern(ra, dec, pol, t)
P = fp**2 + fc**2
A = np.sqrt(P)  # Amplitude antenna pattern

x = A * np.cos(dec) * np.cos(ra)
y = A * np.cos(dec) * np.sin(ra)
z = A * np.sin(dec)


folder = "figures"
os.makedirs(folder, exist_ok=True)

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=A, cmap='viridis', s=2) 

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

cbar = plt.colorbar(sc, ax=ax, shrink=0.4, pad=0.1)
cbar.set_label('Amplitude Antenna Pattern')
plt.tight_layout()
ax.view_init(elev=30, azim=40)

file_path = os.path.join(folder, "amplitude_antenna_pattern.pdf")
plt.savefig(file_path, format='pdf', bbox_inches='tight')

plt.show()
plt.close()
```

## Antenna patterns together

```python
import numpy as np
from pycbc.detector import Detector
import matplotlib.pyplot as plt
import os


n = 100000
ra = np.random.uniform(0, 2 * np.pi, n)
dec = np.pi/2 - np.arccos(np.random.uniform(-1, 1, n))
pol = 0
t = 1e10
folder = "figures"
os.makedirs(folder, exist_ok=True)

# List of detectors and colors
detectors = {
    "L1": "Reds",
    "H1": "Purples"
}

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

for name, cmap in detectors.items():
    d = Detector(name)
    fp, fc = d.antenna_pattern(ra, dec, pol, t)
    P = fp**2 + fc**2

    x = P * np.cos(dec) * np.cos(ra)
    y = P * np.cos(dec) * np.sin(ra)
    z = P * np.sin(dec)

    sc = ax.scatter(x, y, z, c=P, cmap=cmap, alpha=0.2, s=1, label=name)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title("Antenna Patterns of L1, H1")

from matplotlib.lines import Line2D

legend_elements = [
    Line2D([0], [0], marker='o', color='red', label='L1', linestyle='None'),
    Line2D([0], [0], marker='o', color='purple', label='H1', linestyle='None'),]
   

ax.legend(handles=legend_elements, title='Detectors', loc='upper right')

#ax.legend(loc='upper right')
plt.tight_layout()
ax.view_init(elev=30, azim=50)

file_path = os.path.join(folder, "LH_antenna_patterns.png")
plt.savefig(file_path, format='png', bbox_inches='tight')
plt.show()
plt.close()
```
